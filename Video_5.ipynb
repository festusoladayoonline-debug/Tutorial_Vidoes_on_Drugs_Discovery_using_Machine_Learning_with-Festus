# Video 5: Molecular Fingerprints - Converting Molecules to Numbers
Explanation Script:
"Machine learning models don't understand SMILES strings or molecular structures directly. We need to convert molecules into numerical vectors. Molecular fingerprints do exactly this. Think of them as a molecule's unique barcode. RDKit can generate Morgan fingerprints, also called circular fingerprints."

```python
from rdkit import Chem
from rdkit.Chem import AllChem

# Aspirin
smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"
mol = Chem.MolFromSmiles(smiles)

# Generate Morgan fingerprint
fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=1024)

# Convert to bit string and show first 100 bits
fp_string = fp.ToBitString()
print(f"Aspirin fingerprint (first 100 bits):")
print(fp_string[:100])
print(f"Total bits: {len(fp_string)}")
print(f"Number of 1's: {fp_string.count('1')}")
```

"We've converted aspirin into a 1024-bit fingerprint with 24 ones. Each 1 represents a specific molecular substructure present in aspirin. This numerical format is what we feed into machine learning models. The warning is just RDKit informing us about a newer method - our code works fine."
