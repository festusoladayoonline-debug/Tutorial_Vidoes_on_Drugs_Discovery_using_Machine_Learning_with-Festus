# Video 13: Loading Saved ChEMBL Data and Initial Exploration
Explanation Script:
"Now that we've saved our COX-2 data to CSV, let's load it back and explore what we have. This is good practice - we only download once and can reload anytime without hitting the ChEMBL servers again. We'll display the data in a clean table format for better visualization."

```python
import pandas as pd
import numpy as np
from IPython.display import display

# Load the saved data
df = pd.read_csv('cox2_chembl_raw_data.csv')
print(f"Loaded {len(df)} records from 'cox2_chembl_raw_data.csv'")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst 10 records:")

# Display as formatted table
display(df.head(10))

# Basic statistics
print("\n" + "=" * 60)
print("BASIC STATISTICS")
print("=" * 60)
print(f"Total records: {len(df)}")
print(f"Unique molecules: {df['molecule_chembl_id'].nunique()}")
print(f"Unique targets: {df['target_chembl_id'].nunique() if 'target_chembl_id' in df.columns else 'N/A'}")
if 'document_year' in df.columns:
    print(f"Date range: {df['document_year'].min()} - {df['document_year'].max()}")

# Standard value statistics
if 'standard_value' in df.columns:
    print("\nStandard value statistics (nM):")
    print(df['standard_value'].describe())

# Show data types and missing values
print("\n" + "=" * 60)
print("DATA QUALITY CHECK")
print("=" * 60)
missing_df = pd.DataFrame({
    'Column': df.columns,
    'Data Type': df.dtypes.values,
    'Non-Null Count': df.count().values,
    'Null Count': df.isnull().sum().values,
    'Null %': (df.isnull().sum().values / len(df) * 100).round(1)
})
display(missing_df)
```

Explanation of Output:
"The loaded dataset contains 8,383 records with 25 columns. We have 3,892 unique molecules tested against COX-2 between 1975 and 2024. The standard_value column shows IC50 values ranging from 0.1 nM to 450,000 nM. The data quality table helps us identify columns with missing values that need cleaning."
