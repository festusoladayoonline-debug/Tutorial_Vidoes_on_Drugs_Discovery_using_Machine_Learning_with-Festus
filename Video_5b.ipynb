# Video 5b: Comparing Molecular Fingerprints
Explanation Script:
"Let's compare fingerprints of different molecules to understand similarity. Similar molecules should have more bits in common."

```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs

aspirin = Chem.MolFromSmiles("CC(=O)OC1=CC=CC=C1C(=O)O")
ibuprofen = Chem.MolFromSmiles("CC(C)CC1=CC=C(C=C1)C(C)C(=O)O")
benzene = Chem.MolFromSmiles("c1ccccc1")

fp_asp = AllChem.GetMorganFingerprintAsBitVect(aspirin, radius=2, nBits=1024)
fp_ibu = AllChem.GetMorganFingerprintAsBitVect(ibuprofen, radius=2, nBits=1024)
fp_ben = AllChem.GetMorganFingerprintAsBitVect(benzene, radius=2, nBits=1024)

sim1 = DataStructs.TanimotoSimilarity(fp_asp, fp_ibu)
sim2 = DataStructs.TanimotoSimilarity(fp_asp, fp_ben)

print(f"Aspirin vs Ibuprofen: {sim1:.2%}")
print(f"Aspirin vs Benzene: {sim2:.2%}")
```

"Aspirin and ibuprofen are both painkillers with similar structures. Benzene is just a simple ring. Higher similarity means more shared substructures."
