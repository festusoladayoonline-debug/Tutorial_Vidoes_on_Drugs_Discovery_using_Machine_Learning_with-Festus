# Video 5g: Comparing Fingerprints of Different Molecules
Explanation Script:
"Now let's compare which bits are active in aspirin versus ibuprofen. This shows us structural similarities and differences."

```python
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
import numpy as np
import matplotlib.pyplot as plt

aspirin = Chem.MolFromSmiles("CC(=O)OC1=CC=CC=C1C(=O)O")
ibuprofen = Chem.MolFromSmiles("CC(C)CC1=CC=C(C=C1)C(C)C(=O)O")

# Generate fingerprints
bit_info_asp = {}
bit_info_ibu = {}
fp_asp = AllChem.GetMorganFingerprintAsBitVect(aspirin, radius=2, nBits=128, bitInfo=bit_info_asp)
fp_ibu = AllChem.GetMorganFingerprintAsBitVect(ibuprofen, radius=2, nBits=128, bitInfo=bit_info_ibu)

# Convert to arrays
asp_array = np.array(list(fp_asp))
ibu_array = np.array(list(fp_ibu))

print("First 30 bits comparison:")
print("Bit  Aspirin Ibuprofen")
print("-" * 25)
for i in range(30):
    print(f"{i:2}    {asp_array[i]}        {ibu_array[i]}")

# Find common bits
common_bits = [i for i in range(128) if asp_array[i] == 1 and ibu_array[i] == 1]
asp_only = [i for i in range(128) if asp_array[i] == 1 and ibu_array[i] == 0]
ibu_only = [i for i in range(128) if asp_array[i] == 0 and ibu_array[i] == 1]

print(f"\nCommon bits: {common_bits[:10]}...")
print(f"Total common: {len(common_bits)}")
print(f"Aspirin-only bits: {len(asp_only)}")
print(f"Ibuprofen-only bits: {len(ibu_only)}")
```

Explanation Script Continues:
"Both molecules share 12 common bits - these represent substructures present in both, like the benzene ring and carboxylic acid. Aspirin has 8 unique bits not found in ibuprofen, representing its ester group. Ibuprofen has 10 unique bits for its isobutyl chain. Machine learning uses these patterns to learn which combinations of bits correlate with activity."
