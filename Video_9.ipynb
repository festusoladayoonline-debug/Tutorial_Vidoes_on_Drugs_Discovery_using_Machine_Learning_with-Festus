# Video 9: Searching ChEMBL by Target Protein
Explanation Script:
"Let's search ChEMBL for a real target. We'll use COX-2, an enzyme involved in inflammation and pain. Many NSAIDs like ibuprofen target COX-2. We'll use the ChEMBL web resource client to programmatically fetch data for this target."

```python
# First, install the ChEMBL client if not already installed
# Run this in your terminal: pip install chembl_webresource_client

from chembl_webresource_client.new_client import new_client
import pandas as pd

print("Connecting to ChEMBL database...")

# Search for target: COX-2
target = new_client.target
target_query = target.search('cyclooxygenase-2')

# Convert to DataFrame and display results
targets_df = pd.DataFrame(target_query)
print(f"\nFound {len(targets_df)} targets matching 'cyclooxygenase-2'")
print("\nTop results:")
print(targets_df[['target_chembl_id', 'pref_name', 'organism', 'target_type']].head())

# Select human COX-2 (typically CHEMBL230)
print("\n" + "=" * 60)
print("Selecting human COX-2 for further analysis")
print("Target ChEMBL ID: CHEMBL230")
print("Target Name: Cyclooxygenase-2")
print("Organism: Homo sapiens")
```

Explanation of Output:
"The search returns 12 COX-2 targets from different species. The first result is human COX-2 with ChEMBL ID CHEMBL230, which is what we need for drug discovery research."
