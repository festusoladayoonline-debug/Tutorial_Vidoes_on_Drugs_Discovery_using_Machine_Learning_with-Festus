# Video 6: What is QSAR? Introduction to Quantitative Structure-Activity Relationships

Explanation Script:
"QSAR stands for Quantitative Structure-Activity Relationship. It's a computational method that correlates chemical structure with biological activity. The fundamental principle is simple: molecules with similar structures tend to have similar biological effects. QSAR has been used since the 1960s to predict drug properties before synthesis, saving time and resources. Let's see a simple example using molecular weight as a proxy for activity."

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
import matplotlib.pyplot as plt
import numpy as np

# Sample molecules with known activities (simulated)
molecules = [
    ("Aspirin", "CC(=O)OC1=CC=CC=C1C(=O)O", 1),
    ("Ibuprofen", "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O", 1),
    ("Caffeine", "CN1C=NC2=C1C(=O)N(C(=O)N2C)C", 0),
    ("Paracetamol", "CC(=O)NC1=CC=C(O)C=C1", 1),
    ("Benzene", "c1ccccc1", 0),
    ("Ethanol", "CCO", 0),
    ("Toluene", "CC1=CC=CC=C1", 0),
    ("Naphthalene", "C1=CC=C2C=CC=CC2=C1", 0)
]

# Calculate molecular weight for each
names = []
weights = []
activities = []

for name, smiles, active in molecules:
    mol = Chem.MolFromSmiles(smiles)
    mw = Descriptors.MolWt(mol)
    names.append(name)
    weights.append(mw)
    activities.append(active)
    
    status = "ACTIVE" if active == 1 else "inactive"
    print(f"{name:12} MW: {mw:5.1f} - {status}")

# Simple visualization
plt.figure(figsize=(8, 4))
colors = ['green' if a == 1 else 'red' for a in activities]
plt.scatter(weights, activities, c=colors, s=100)
plt.xlabel('Molecular Weight')
plt.ylabel('Activity (1=Active, 0=Inactive)')
plt.title('QSAR Concept: Structure vs Activity')
plt.yticks([0, 1], ['Inactive', 'Active'])
plt.grid(True, alpha=0.3)
plt.show()

print("\nObserve: Active drugs cluster around MW 150-210")
print("This is a simple QSAR relationship - weight influences activity")
```

Explanation of Output:
"The output shows that active drugs cluster in the 150-210 molecular weight range while inactive molecules are mostly outside this range. This demonstrates the basic QSAR principle that molecular weight correlates with activity."
