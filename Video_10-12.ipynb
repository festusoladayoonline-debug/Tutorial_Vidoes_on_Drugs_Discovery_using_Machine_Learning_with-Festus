# Video 10-12: Downloading ChEMBL Data for COX-2
Explanation Script:
"Let's download all bioactivity data for human COX-2. We'll fetch compounds tested against this target, including their structures and IC50 values. After downloading, we'll save the data to a CSV file so we don't have to re-download it every time we work on this project."

```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
from IPython.display import display

# Get activities for COX-2 (CHEMBL230)
activities = new_client.activity
target_chembl_id = 'CHEMBL230'

print(f"Fetching bioactivity data for target {target_chembl_id}...")
res = activities.filter(target_chembl_id=target_chembl_id).filter(standard_type="IC50")

# Convert to DataFrame
df = pd.DataFrame(res)
print(f"Retrieved {len(df)} bioactivity records")

# Check available columns
print("\nAvailable columns in the data:")
print(df.columns.tolist())
print(f"\nTotal columns: {len(df.columns)}")

# Select columns that actually exist
available_cols = df.columns.tolist()
relevant_cols = ['molecule_chembl_id', 'standard_value', 'standard_units', 
                 'assay_description', 'document_year', 'canonical_smiles']

# Only keep columns that are in the dataframe
existing_cols = [col for col in relevant_cols if col in available_cols]
print(f"\nSelecting these available columns: {existing_cols}")

df_subset = df[existing_cols].copy()
print("\nSample records (first 10 rows):")
display(df_subset.head(10))

print(f"\nTotal records: {len(df)}")
print(f"Unique molecules: {df['molecule_chembl_id'].nunique()}")
if 'document_year' in df.columns:
    print(f"Date range: {df['document_year'].min()} - {df['document_year'].max()}")

# Show data summary
print("\n" + "=" * 60)
print("DATA SUMMARY")
print("=" * 60)
summary_df = pd.DataFrame({
    'Metric': ['Total Records', 'Unique Molecules', 'Date Range', 'Columns'],
    'Value': [
        f"{len(df)}", 
        f"{df['molecule_chembl_id'].nunique()}",
        f"{df['document_year'].min()} - {df['document_year'].max()}" if 'document_year' in df.columns else 'N/A',
        f"{len(df.columns)}"
    ]
})
display(summary_df)

# Save the full raw data to CSV
df.to_csv('cox2_chembl_raw_data.csv', index=False)
print(f"\n✓ Data saved to 'cox2_chembl_raw_data.csv'")
print(f"✓ File contains {len(df)} records with {len(df.columns)} columns")
```

Explanation of Output:
"We successfully retrieved 8,383 bioactivity records for COX-2 from 6,349 unique molecules. The data spans from 1994 to 2024 and includes SMILES structures. The complete dataset is now saved as 'cox2_chembl_raw_data.csv' in your working directory for future use without re-downloading."
