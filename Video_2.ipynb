# Video 2: What's a Molecule? SMILES Strings Explained

Explanation Script:
"Computers don't understand chemical structures directly. Instead, we use SMILES â€“ a text representation.
For example, aspirin's SMILES is CC(=O)OC1=CC=CC=C1C(=O)O.
Let's use RDKit to convert that SMILES into a molecule object and visualize it.
Run this code in Jupyter:"

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Aspirin SMILES
aspirin_smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"

# Create molecule object
mol = Chem.MolFromSmiles(aspirin_smiles)

# Draw the molecule (image will appear in notebook)
Draw.MolToImage(mol)
```
