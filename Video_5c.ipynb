# Video 5c: Fingerprint Bits as Features
Explanation Script:
"Each fingerprint bit becomes a feature for machine learning. Let's see how to convert multiple molecules into a feature matrix."

```python
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np

smiles_list = [
    "CC(=O)OC1=CC=CC=C1C(=O)O",  # Aspirin
    "CN1C=NC2=C1C(=O)N(C(=O)N2C)C",  # Caffeine
    "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"  # Ibuprofen
]

fingerprints = []
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=128)
    fingerprints.append(list(fp))

X = np.array(fingerprints)
print(f"Feature matrix shape: {X.shape}")
print(f"Molecules: {X.shape[0]}, Features: {X.shape[1]}")
print("\nFirst molecule first 20 bits:")
print(X[0, :20])
```

"We now have a feature matrix ready for machine learning. Each row is a molecule, each column is a fingerprint bit. This is what we'll use to train our models."
