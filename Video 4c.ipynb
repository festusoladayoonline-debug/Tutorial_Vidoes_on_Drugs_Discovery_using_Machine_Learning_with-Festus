Hydrogen Bond Acceptors - Second Lipinski Rule
Explanation Script: "The second Lipinski rule states hydrogen bond acceptors should be less than 10. Acceptors are typically oxygen and nitrogen atoms. Let's calculate this for our drug set."

```python
from rdkit import Chem
from rdkit.Chem import Lipinski

# Aspirin
smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"
mol = Chem.MolFromSmiles(smiles)
hba = Lipinski.NumHAcceptors(mol)
print(f"Aspirin H-bond acceptors: {hba}")

# Caffeine
caffeine_smiles = "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"
mol_caf = Chem.MolFromSmiles(caffeine_smiles)
hba_caf = Lipinski.NumHAcceptors(mol_caf)
print(f"Caffeine H-bond acceptors: {hba_caf}")

# Paracetamol
para_smiles = "CC(=O)NC1=CC=C(O)C=C1"
mol_para = Chem.MolFromSmiles(para_smiles)
hba_para = Lipinski.NumHAcceptors(mol_para)
print(f"Paracetamol H-bond acceptors: {hba_para}")
```

"We've calculated hydrogen bond acceptors. All are under 10, so these drugs pass the second rule as well."
