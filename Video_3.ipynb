# Video 3: Visualizing Multiple Molecules at Once
Explanation Script: "Let's visualize several molecules in a grid. This is great for comparing structures. We'll create a list of SMILES for common drugs and display them side by side."

```python
from rdkit import Chem
from rdkit.Chem import Draw

# List of SMILES (aspirin, caffeine, ibuprofen, ethanol)
smiles_list = [
    "CC(=O)OC1=CC=CC=C1C(=O)O",  # aspirin
    "CN1C=NC2=C1C(=O)N(C(=O)N2C)C",  # caffeine
    "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O",  # ibuprofen
    "CCO"  # ethanol
]

# Convert to molecule objects
mols = [Chem.MolFromSmiles(s) for s in smiles_list]

# Draw grid
img = Draw.MolsToGridImage(mols, molsPerRow=2, subImgSize=(200,200))
img
```
