# Video 5e: What Substructure Does Each Bit Represent?
Explanation Script:
"RDKit can tell us exactly which atom environments create each '1' in the fingerprint. Let's map bit 2 and bit 5 to actual molecular fragments."

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import SimilarityMaps
import matplotlib.pyplot as plt

# Aspirin
smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"
mol = Chem.MolFromSmiles(smiles)

# Generate fingerprint with bit info
fp = AllChem.GetMorganFingerprint(mol, radius=2)
bit_info = {}

# Get bit 2 details
bit_2_fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=128, bitInfo=bit_info)
print("Bit 2 appears in these atom environments:")
if 2 in bit_info:
    for atom, radius in bit_info[2]:
        print(f"  - Centered on atom {atom} with radius {radius}")
        
        # Get the environment
        env = Chem.FindAtomEnvironmentOfRadiusN(mol, radius, atom)
        amap = {}
        submol = Chem.PathToSubmol(mol, env, atomMap=amap)
        print(f"    Substructure atoms: {sorted(amap.keys())}")

# Get bit 5 details
print("\nBit 5 appears in these atom environments:")
if 5 in bit_info:
    for atom, radius in bit_info[5]:
        print(f"  - Centered on atom {atom} with radius {radius}")
```

Explanation Script Continues:
"This shows that bit 2 = 1 because aspirin has two environments: one around atom 4 and one around atom 8, each with radius 1 (the atom and its immediate neighbors). Bit 5 = 1 because of a larger environment around atom 0 with radius 2. So each bit doesn't represent just one thing - it activates whenever ANY environment in the molecule matches that specific pattern. This is why fingerprints are powerful: they capture all instances of each substructure type."
