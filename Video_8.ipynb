Explanation Script:
"ChEMBL is an open database containing bioactivity data for millions of compounds. It's maintained by the European Bioinformatics Institute and includes data from medicinal chemistry literature, patents, and drug approvals. Each entry includes compound structure, target information, assay details, and measured activity. Let's explore what ChEMBL offers using their web interface and understand the data structure."

```python
# This video introduces ChEMBL - we'll use their API in next video
# For now, let's simulate what ChEMBL data looks like

import pandas as pd

# Simulated ChEMBL-like data structure
chembl_data = {
    'ChEMBL_ID': ['CHEMBL25', 'CHEMBL108', 'CHEMBL120', 'CHEMBL521', 'CHEMBL539'],
    'Compound_Name': ['ASPIRIN', 'IBUPROFEN', 'CAFFEINE', 'PARACETAMOL', 'NAPROXEN'],
    'SMILES': [
        'CC(=O)OC1=CC=CC=C1C(=O)O',
        'CC(C)CC1=CC=C(C=C1)C(C)C(=O)O',
        'CN1C=NC2=C1C(=O)N(C(=O)N2C)C',
        'CC(=O)NC1=CC=C(O)C=C1',
        'CC(C1=CC=C(C=C1)OC)C(=O)O'
    ],
    'Target': ['COX-1', 'COX-1', 'Adenosine A2a', 'COX-2', 'COX-2'],
    'Assay_Type': ['Binding', 'Binding', 'Functional', 'Binding', 'Binding'],
    'IC50_nM': [5000, 4500, 21000, 18000, 8500],
    'Confidence_Score': [9, 8, 9, 7, 8],
    'Publication_Year': [1995, 1987, 2001, 1993, 1982]
}

df = pd.DataFrame(chembl_data)
print("ChEMBL Data Structure Example:")
print("=" * 70)
print(df.to_string(index=False))

print("\n" + "=" * 70)
print(f"Total compounds: {len(df)}")
print(f"Data fields: {', '.join(df.columns)}")
print("\nEach record contains:")
print("- Compound identifier and structure")
print("- Target protein information")
print("- Assay conditions and results")
print("- Confidence in the data")
print("- Source publication")
```

Explanation of Output:
"The output shows a sample of ChEMBL data structure with five drug molecules, each containing their ChEMBL ID, chemical structure, target protein, assay type, IC50 value, confidence score, and publication year. This is the raw format we work with."
