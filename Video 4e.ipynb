# Video 4e: Molecular Weight - Fourth Lipinski Rule
Explanation Script:
"The fourth Lipinski rule requires molecular weight under 500. Larger molecules struggle to cross cell membranes. Let's check our drugs."

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

drugs = [
    ("Aspirin", "CC(=O)OC1=CC=CC=C1C(=O)O"),
    ("Ibuprofen", "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"),
    ("Caffeine", "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"),
    ("Paracetamol", "CC(=O)NC1=CC=C(O)C=C1")
]

for name, smiles in drugs:
    mol = Chem.MolFromSmiles(smiles)
    mw = Descriptors.MolWt(mol)
    print(f"{name}: {mw:.1f} g/mol")
```

"We've calculated molecular weights. All are well under 500, confirming these are drug-like molecules."
