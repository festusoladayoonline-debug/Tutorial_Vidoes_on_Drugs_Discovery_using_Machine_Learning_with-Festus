# Video 4b: Hydrogen Bond Donors - First Lipinski Rule
Explanation Script: "Lipinski's Rule of Five helps predict if a molecule can be an oral drug. The first rule: hydrogen bond donors should be less than 5. H-bond donors are hydrogens attached to oxygen or nitrogen. Let's calculate this for common drugs using RDKit's Lipinski module."

```python
from rdkit import Chem
from rdkit.Chem import Lipinski

# Aspirin
smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"
mol = Chem.MolFromSmiles(smiles)
hbd = Lipinski.NumHDonors(mol)
print(f"Aspirin H-bond donors: {hbd}")

# Caffeine
caffeine_smiles = "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"
mol_caf = Chem.MolFromSmiles(caffeine_smiles)
hbd_caf = Lipinski.NumHDonors(mol_caf)
print(f"Caffeine H-bond donors: {hbd_caf}")

# Ibuprofen
ibuprofen_smiles = "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"
mol_ibu = Chem.MolFromSmiles(ibuprofen_smiles)
hbd_ibu = Lipinski.NumHDonors(mol_ibu)
print(f"Ibuprofen H-bond donors: {hbd_ibu}")
```
